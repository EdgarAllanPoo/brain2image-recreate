```python
import numpy as np
import pickle
```

```python
with open('data/eeg/char/data.pkl', 'rb') as file:
	data = pickle.load(file, encoding='latin1')
	train_X = data['x_train']
	train_Y = data['y_train']
	test_X = data['x_test']
	test_Y = data['y_test']
```

```python
print(train_X.shape)
print(train_X[0].shape)
print(train_X[0][0].shape)
```

```python
print(type(train_X[0][0]))
print(train_X[0][0][0][0])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

instance_a = train_X[0]

time_points = np.arange(32)  # Waktu dari 0 hingga 31

channel_names = ["AF3", 
                 "F7", 
                 "F3", 
                 "FC5", 
                 "T7",
                 "P7",
                 "O1",
                 "O2",
                 "P8",
                 "T8",
                 "FC6",
                 "F4",
                 "F8",
                 "AF4"]

plt.figure(figsize=(10, 6))

for i in range(14):
    plt.plot(time_points, instance_a[i], label=f'{channel_names[i]}')

plt.xlabel('Time')
plt.ylabel('Potential')
plt.title('Values over Time for 14 Channels')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```

```python
instance_b = train_X[1]

time_points = np.arange(32)  # Waktu dari 0 hingga 31

plt.figure(figsize=(10, 6))

for i in range(14):
    plt.plot(time_points, instance_b[i], label=f'Channels {i+1}')

plt.xlabel('Time')
plt.ylabel('Potential')
plt.title('Values over Time for 14 Channels')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```
